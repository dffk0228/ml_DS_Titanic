# Data Preprocessing

## Mengisi data kosong pada Kolom Age dengan Median

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df
```

Banyak cara untuk mengisi data kosong, bisa hapus baris yang kosong, bisa isi dengan rata-rata, median (angka yang di tengah). Tetapi mengisi data kosong dengan median terkadang lebih sering digunakan ketimbang dengan rata-rata karena rata-rata bisa mengambil data outliers/pencilan sehingga lebih disarankan untuk median saja (nilai tengah).

## Isi data kosong pada Kolom Embarked dengan Modus

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df
```

1. Mengapa perlu [0]? Dalam kasus ini jika ada lebih dari satu modus (misalnya, jika "S" dan "C" muncul dengan frekuensi yang sama), df['Embarked'].mode() akan mengembalikan lebih dari satu nilai. [0] memastikan bahwa Anda hanya mengambil modus pertama dalam daftar.

2. df['Embarked'].mode()[0] mengambil nilai modus dari kolom "Embarked" dan menggunakan nilai tersebut untuk mengisi nilai yang hilang (NaN) dalam kolom tersebut.

3. Misalkan jika modus dari embarked pada data ternyata 2 yaitu S dan C. Berarti terdapat 2 modus. Yang diambil adalah modus pertama dalam daftar aja, tidak usah dua duanya.

Kolom Pclass dan Survive diubah tipe datanya dari integer (bilangan bulat) ke string agar kompatibel dengan fungsi seaborn/matplotlib 

```python
import matplotlib.pyplot as plt
# Convert the 'Survived' column back to numeric
df['Survived'] = df['Survived'].astype(int)

colors = ['#FF9999', '#66B2FF', '#99FF99']  # Warna berbeda untuk setiap kelas

# Plot manual untuk setiap kelas
classes = df['Pclass'].unique()
for idx, cls in enumerate(classes):
    subset = df[df['Pclass'] == cls]
    sns.barplot(x=[str(cls)], y=[subset['Survived'].mean()], color=colors[idx])

# Plotting data
sns.barplot(data=df, x='Pclass', y='Survived')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

## Feature Selection

Untuk memprediksi keselamatan penumpang Titanic, model akan dilatih untuk memprediksi apakah seorang penumpang selamat (y) berdasarkan fitur-fitur seperti kelas tiket (Pclass), jenis kelamin (Sex), usia (Age), jumlah anggota keluarga (SibSp & Parch), tarif tiket (Fare), dan pelabuhan keberangkatan (Embarked) (X).

```python
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
X = df[features] #predictor
y = df['Survived'] #target
```

1. 'Features' adalah kolom-kolom yang dipilih untuk digunakan sebagai input model machine learning.
2.  X adalah DataFrame baru yang berisi kolom-kolom yang ada dalam daftar 'features'. Tujuannnya untuk membuat Matriks Fitur.
3.  Matriks Fitur: Matriks ini adalah kumpulan data yang akan digunakan sebagai input untuk model machine learning. Setiap baris dalam X mewakili satu penumpang, dan setiap kolom mewakili satu fitur atau karakteristik dari penumpang tersebut.

1. Y adalah target atau label yang ingin diprediksi oleh model
2. Kolom 'Survived' berisikan nilai biner 0 dan 1 (0 penumpang tidak selamat, 1 penumpang selamat)
3. Y berisikan hasil atau keluaran yang diinginkan untuk setiap baris X.
4. Y digunakan sebagai referensi untuk melatih model machine learning, sehingga model dapat mempelajari pola-pola dari X untuk memprediksi y

# Data Modeling

## Membangun classification model menggunakan Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

- n_estimators = 100, parameter ini menunjukkan jumlah decision trees (pohon keputusan) yang akan dibangun oleh model ini. Dalam model ini berarti akan dibangun 100 pohon keputusan
- random_state = 42, sama seperti 'train_test_split', 'random_state' digunakan untuk memastikan bahwa hasil pelatihan dapat diprediksi. 'random_state' memastikan bahwa pemilihan acak dalam proses pembentukan pohon keputusan akan sama setiap kali kode dijalankan.
- Melatih model dalam data pelatihan (model.fit(X_train, y_train)):
  Model akan mempelajari pola-pola dalam data sehingga dapat membuat prediksi pada data baru.

# Evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

## Prediksi data test menggunakan model random forest yang sudah dibangun

```python
y_pred = model.predict(X_test)
```

```python
df_comparison = pd.DataFrame({
    'Actual': y_test,
    'Predicted': y_pred
})
df_comparison
```

## Membuat confusion matrix

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

Untuk membandingkan antara test dengan model.


## Menghitung akurasi dan prediksi

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```
